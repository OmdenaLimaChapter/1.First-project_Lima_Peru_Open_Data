- Department: where the Payment was made 
- Province: where the Payment was made 
- District: where the Payment was made 
- Geographic: location code where the payment was made 
- Date of payment 
- Id that identifies the taxpayer (anonymized) 
- Person to whom payment is assigned (anonymized) 
- Payment month 
- Payment concept 
- Years of accounts paid [*Expressed in Peruvian Sol currency]
- Total paid per row [*Expressed in Peruvian Sol currency]

```python
# Assuming PAYMENT_DATE is in YYYYMMDD format
df2['PAYMENT_DATE'] = pd.to_datetime(df2['PAYMENT_DATE'], format='%Y%m%d')
df2['PAYMENT_DATE'] = df2['PAYMENT_DATE'].dt.date

# Assuming MONTH is in YYYYMM format
df2['MONTH'] = pd.to_datetime(df2['MONTH'], format='%Y%m').dt.to_period('M')

# Display the updated DataFrame
print(df2[['PAYMENT_DATE', 'MONTH']].head())
```

## Time Series Analysis:

### (i) Analyze trends in total payments over time:

```python
# Create a new DataFrame for resampling
df_resample = df2.groupby('MONTH')['TOTAL'].sum()

# Plot the time series of total payments
plt.figure(figsize=(8, 4))
sns.lineplot(x=df_resample.index.astype(str), y=df_resample.values, marker='o')
plt.title('Total Payments Over Time')
plt.xlabel('Time (Months)')
plt.ylabel('Total Payments (Peruvian Sol)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

There appears to be some seasonality in the data, with noticeable fluctuations in payment amounts throughout different months. 


```python
# (ii) Examine changes in payment amounts for specific concepts over time:

# Select specific payment concepts for analysis
selected_concepts = ['IMPUESTO PREDIAL', 'LIMPIEZA', 'SERENAZGO']

# Filter the DataFrame for the selected concepts
df_selected_concepts = df2[df2['CONCEPT'].isin(selected_concepts)]

# Pivot the DataFrame for better visualization
df_selected_concepts_pivot = df_selected_concepts.pivot_table(values='TOTAL', index='MONTH', columns='CONCEPT', aggfunc='sum')

# Plot the time series of payment amounts for specific concepts
plt.figure(figsize=(12, 6))
df_selected_concepts_pivot.plot.line(marker='o')
plt.title('Payment Amounts for Specific Concepts Over Time')
plt.xlabel('Time (Months)')
plt.ylabel('Payment Amount (Peruvian Sol)')
plt.legend(title='Concept')
plt.grid(True)
plt.tight_layout()
plt.show()
```

All concepts show increasing trends, with property tax having the highest overall amount and security the lowest.
Property tax exhibits the most significant fluctuations, while cleaning payments are relatively stable.

## Conceptual Analysis:

```python
# Investigate the most common payment concepts and their contribution to total revenue.
# Group the data by payment concept and calculate the sum of total payments for each concept
concept_totals = df2.groupby('CONCEPT')['TOTAL'].sum()

# Sort the concepts by total revenue in descending order
concept_totals = concept_totals.sort_values(ascending=False)

# Plot the contribution of each concept to total revenue
plt.figure(figsize=(12, 6))
sns.barplot(x=concept_totals.index, y=concept_totals.values)
plt.title('Contribution of Payment Concepts to Total Revenue')
plt.xlabel('Payment Concept')
plt.ylabel('Total Revenue (Peruvian Sol)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

The most common payment concepts are PROPERTY TAX, CLEANING, FEES, PARKS AND GARDENS, MUNICIPAL SECURITY, and ADMINISTRATIVE FINE.

The remaining payment concepts are less common and contribute less to total revenue.


##  Correlation Analysis:

```python
# Explore correlations between different variables
correlation_matrix = df2[['2022', '2021', '2020', '2019', '2018', 'TOTAL']].corr()

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(8, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Payment Amounts')
plt.show()
```

**Strong positive correlations:**
There are strong positive correlations between all years, especially between adjacent years (e.g., 2021 and 2022, 2020 and 2021). This suggests that payment amounts in one year tend to be similar to the amounts in the following year.
The total payment amount also shows strong positive correlations with all years, indicating that higher overall payment amounts in one year are associated with higher amounts in individual years as well. \
**Weaker correlations:**
The correlations between 2018 and the other years are slightly weaker than the correlations between later years. This could suggest that payment patterns have changed somewhat over time.

## 3.2. Temporal Analysis: 
#### Are there specific months that consistently show higher or lower income collection?

```python
# Identify specific months with higher or lower income collection
df2.set_index('MONTH', inplace=True)

monthly_income_distribution = df2.groupby(df2.index.month)['TOTAL'].sum()

plt.figure(figsize=(12, 6))
monthly_income_distribution.plot(kind='bar', color='blue')
plt.title('Monthly Income Distribution')
plt.xlabel('Month')
plt.ylabel('Total Income Collection (Peruvian Sol)')
plt.show()
```

**Seasonality:**
- Higher income months: The graph clearly shows higher income collection in the months of April, May, June, and July. This could be due to factors like: 
  - Seasonal work: Some industries like agriculture or tourism might have peak seasons during these months, leading to higher income for workers.
  - Bonuses: Certain companies might distribute annual or performance-based bonuses around this time, boosting overall income.
  - Tax filing: April is the deadline for tax filing in Peru, potentially leading to temporary increases in income reporting due to refunds or adjustments.

- Lower income months: The graph shows lower income collection in December, January, and February. This could be due to: 
  - Holidays: December and January are major holiday months in Peru, with reduced economic activity and potential temporary closures of businesses.
  - School holidays: January and February also coincide with school holidays, potentially impacting income in education-related sectors.
  - Post-holiday spending: January might see increased spending after the holidays, leading to a temporary decrease in savings or reported income.
