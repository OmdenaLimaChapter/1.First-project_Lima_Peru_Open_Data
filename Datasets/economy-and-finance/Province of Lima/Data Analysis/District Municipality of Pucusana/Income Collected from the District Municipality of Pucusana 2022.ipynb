### Dataset Description :
The information corresponds to all income collected by the District Municipality of Pucusana in the year 2022 (January - September). The types of income correspond to property tax, taxes (public cleaning, parks and gardens or serenazgo) and other concepts such as: forms, cards, certificates, fines, etc. The location of this dataset is the Department of Lima, Province of Lima, District of Pucusana with Ubigeo 140118. 

#### The dataset is characterized by:
Movement number, type of movement, person's code. Personal data such as documents, names, surnames and address were correctly anonymized in compliance with current regulations. The amounts collected include the type of payment, property location, concept, cancellation date, among other data. Open data is important to contribute to the transparency of municipal management and the openness of data.

## Objective:
To gain insights into the revenue collected by the District Municipality of Pucusana in Lima Province for the year 2022 (Jan to Sept). By exploring the patterns and details within the dataset, the aim is to better understand the financial dynamics, identify trends, and derive actionable insights to contribute to the improvement of economy and quality of life.

## Revenue Analysis[Concept-specific Revenue Distribution]:

```python
# 1. Analyze the distribution of revenue across different concepts
sns.set(style="whitegrid") # Set the style for seaborn
plt.figure(figsize=(8, 4))
sns.boxplot(x='CONCEPT_D', y='TOTAL_COLLECTED', data=df)
plt.title('Distribution of Revenue Across Different Concepts')
plt.xlabel('Concepts')
plt.ylabel('Total Collected Amount')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# 2. Identify the top revenue-generating concepts and their contribution to the total income.
top_concepts = df.groupby('CONCEPT_D')['TOTAL_COLLECTED'].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(8, 4))
top_concepts.plot(kind='bar', color='skyblue')
plt.title('Top Revenue-Generating Concepts')
plt.xlabel('Concepts')
plt.ylabel('Total Collected Amount')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Public cleaning is the most important source of revenue for the municipality, generating almost half of the total revenue from the top five concepts.
The other four concepts Security, Property Tax, Parks and Gardens are all relatively close in terms of revenue generated.

```python
# 3. Identify the Revenue distribution for Top Observations
top_observation_revenue = df.groupby('OBSERVATIONS')['TOTAL_COLLECTED'].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(8, 4))
top_observation_revenue.plot(kind='bar', color='orange')
plt.title('Revenue Distribution for Top Observations')
plt.xlabel('Observations')
plt.ylabel('Total Collected Amount')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Property Tax and Municipal Taxes stand out as the highest observations for total collected amounts.
Municipal Taxes is the second-highest observation, albeit relatively lower than the top observation.
Property Tax, Arbitrators, and Property Tax, Municipal Taxes, Tax Penalties are also notable observations.

## Temporal Analysis:

```python
# 1. Explore temporal trends by analyzing how revenue collection has changed over the months 2022.
monthly_revenue = df.groupby(df['CANCEL_DATE'].dt.to_period("M"))['TOTAL_COLLECTED'].sum()

# Plotting the temporal trends
plt.figure(figsize=(8, 4))
monthly_revenue.plot(marker='o', linestyle='-')
plt.title('Temporal Trends in Revenue Collection')
plt.xlabel('Month')
plt.ylabel('Total Collected Amount')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Revenue collection fluctuated throughout the months of January to September 2022.
Month with the Highest Revenue is Sept 2022 and
Month with the Lowest Revenue is Oct 2022

```python
# 2. Are there particular months or quarters where the municipality tends to collect more revenue?

# Extracting month and quarter from the 'CANCEL_DATE' column
df['Month'] = df['CANCEL_DATE'].dt.month
df['Quarter'] = df['CANCEL_DATE'].dt.quarter

# Grouping by month and quarter, and summing the total collected amount for each
monthly_trends = df.groupby('Month')['TOTAL_COLLECTED'].sum()
quarterly_trends = df.groupby('Quarter')['TOTAL_COLLECTED'].sum()

# Plotting the temporal trends
plt.figure(figsize=(16, 6))

plt.subplot(1, 2, 1)
monthly_trends.plot(kind='bar', color='skyblue')
plt.title('Monthly Trends in Revenue Collection')
plt.xlabel('Month')
plt.ylabel('Total Collected Amount')

plt.subplot(1, 2, 2)
quarterly_trends.plot(kind='bar', color='lightcoral')
plt.title('Quarterly Trends in Revenue Collection')
plt.xlabel('Quarter')
plt.ylabel('Total Collected Amount')

plt.tight_layout()
plt.show()
```

Quarterly analysis reveals that Quarter 3 exhibits the highest revenue, followed by Quarters 1 and 2, with Quarter 4 recording the lowest revenue.

 ## Correlation Analysis:

## Impact of Discounts: 

```python
# How do offered discounts impact the total collected amount, and is there a noticeable correlation between discounts and timely payments?

# Scatter plot to visualize the relationship between discounts and total collected amount
plt.figure(figsize=(8, 4))
sns.scatterplot(x='DISCOUNT_D', y='TOTAL_COLLECTED', data=df)
plt.title('Impact of Discounts on Total Collected Amount')
plt.xlabel('Discount Amount')
plt.ylabel('Total Collected Amount')
plt.show()

# Correlation analysis
discount_total_corr = df['DISCOUNT_D'].corr(df['TOTAL_COLLECTED'])
print(f"Correlation between Discounts and Total Collected Amount: {discount_total_corr}")
```

The correlation between discounts and the total collected amount is 0.096, indicating a weak positive correlation. This suggests that offered discounts have a slight positive impact on the total collected amount.

## 3. Efficiency in Revenue Collection:

```python
# Calculate potential revenue for each concept
potential_revenue_per_concept = df.groupby('CONCEPT_D')['AMOUNT_D'].sum()

# Calculate actual collected revenue for each concept
collected_revenue_per_concept = df.groupby('CONCEPT_D')['TOTAL_COLLECTED'].sum()

# Calculate efficiency ratio for each concept
efficiency_ratio_per_concept = collected_revenue_per_concept / potential_revenue_per_concept

# Display the efficiency of revenue collection for each concept
print("Efficiency of Revenue Collection for Each Concept:")
print(efficiency_ratio_per_concept)
```

```python
# Identify concepts with lower efficiency (below a certain threshold, e.g., 0.8)
lower_efficiency_concepts = efficiency_ratio_per_concept[efficiency_ratio_per_concept < 0.8]

# Display concepts with lower efficiency
print("Concepts with Lower Efficiency and Areas for Improvement:")
print(lower_efficiency_concepts)
```

```python
# Calculate potential revenue for each payment type
potential_revenue_per_payment_type = df.groupby('PAYMENT_TYPE')['AMOUNT_D'].sum()

# Calculate actual collected revenue for each payment type
collected_revenue_per_payment_type = df.groupby('PAYMENT_TYPE')['TOTAL_COLLECTED'].sum()

# Calculate efficiency ratio for each payment type
efficiency_ratio_per_payment_type = collected_revenue_per_payment_type / potential_revenue_per_payment_type

# Display the efficiency of revenue collection for each payment type
print("Efficiency of Revenue Collection for Each Payment Type:")
print(efficiency_ratio_per_payment_type)
```
