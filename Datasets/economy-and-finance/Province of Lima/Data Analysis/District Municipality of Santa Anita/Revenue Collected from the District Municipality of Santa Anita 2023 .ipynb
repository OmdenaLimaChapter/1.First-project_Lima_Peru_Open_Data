### Dataset Description :
Data on Tax Collection Income of the District Municipality of Santa Anita corresponding to the year 2023 (January to May).The types of income correspond to Property Tax, Excise Taxes (Public Cleaning, Parks and Gardens or Serenazgo) and other concepts such as: forms, cards, certificates, fines, etc.This information is important to contribute to the transparency of municipal management and the openness of data.

#### The dataset is characterized by: 
- Ubigeo: Geographic location code where the payment was made.
- Department: Department where the Payment was made. 
- Province: Province where the Payment was made. 
- District: District where the Payment was made. 
- Citizen Data: Id that identifies the taxpayer (anonymized) 
- Type of Tax : Type of Tax collected, in this case property tax. 
- Years paid: The years that remained to be paid to the citizen. 
- Number of properties: Number of paid real estate possessions. 
- Property Location: Reference location of the property. 
- Tax: Tax paid. 
- Default interest: Interest for late payment by the taxpayer. 
- Balance: Balance that remains to be paid. 
- Date: Date of payment. 
- Receipt: Receipt number. 
- Payment: Total paid per row [*Expressed in Peruvian Sol currency*]

## Revenue Distribution Analysis:

```python
# 1. Explore the distribution of revenue across different types of concepts
# Identify the top 10 concepts
top_10_concepts = df.groupby('CONCEPT')['TOTAL_COLLECTED'].sum().nlargest(10).sort_values().index

# Filter the DataFrame for the top 10 concepts
df_top_10 = df[df['CONCEPT'].isin(top_10_concepts)]

# Explore the distribution of revenue across different types of taxes and concepts (Top 10)
plt.figure(figsize=(14, 8))
sns.set(style="whitegrid")
sns.barplot(x='TOTAL_COLLECTED', y='CONCEPT', data=df_top_10, ci=None)
plt.title('Distribution of Revenue Across Top 10 Concepts')
plt.xlabel('Total Collected Amount (Peruvian Sol)')
plt.ylabel('Concept')
plt.show()
```

The top revenue-generating concepts include taxes on non-sporting public spectacles, medical care, garbage collection, parks and gardens, street sweeping, Municipal Security, property tax, administrative control fines, subdivision-related payments, and public cleaning. These concepts indicate a diverse range of revenue sources, with fines and property-related taxes playing a significant role.

```python
# 2. Identify the Revenue distribution for Top Observations

# Group by 'OBSERVATIONS' and sum the 'TOTAL_COLLECTED' for each observation
top_observations = df.groupby('OBSERVATIONS')['TOTAL_COLLECTED'].sum().sort_values(ascending=True).head(10)

# Filter the DataFrame for the top observations
df_top_observations = df[df['OBSERVATIONS'].isin(top_observations.index)]

# Plot the distribution of revenue across top observations
plt.figure(figsize=(14, 8))
sns.set(style="whitegrid")
sns.barplot(x='TOTAL_COLLECTED', y='OBSERVATIONS', data=df_top_observations, ci=None, palette='viridis')
plt.title('Revenue Distribution for Top Observations')
plt.xlabel('Amount (Peruvian Sol)')
plt.ylabel('Observations')
plt.show()
```

Observations such as "simple copies," additional receipts, and specific references like "pool universal 19" have minimal impact on revenue, each contributing less than 2% to the total collected amount.

```python
# 3. Explore the distribution of payment methods and their impact on revenue collection
payment_method_distribution = df['PAYMENT_TYPE'].value_counts()
total_collected_per_payment_method = df.groupby('PAYMENT_TYPE')['TOTAL_COLLECTED'].sum()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

payment_method_distribution.plot(kind='bar', ax=ax1)
ax1.set_title('Distribution of Payment Methods')
ax1.set_xlabel('Payment Method')
ax1.set_ylabel('Number of Transactions')

total_collected_per_payment_method.plot(kind='bar', ax=ax2)
ax2.set_title('Total Revenue Collected by Payment Method')
ax2.set_xlabel('Payment Method')
ax2.set_ylabel('Total Revenue (in Peruvian Sol)')

plt.show()
```

The municipality primarily receives revenue through 'CASH', credit card payments ('VISA'), and 'TRANSFERS'. While cash transactions are numerous, transfers constitute the highest total revenue.

## Temporal Analysis:

```python
# 1. Explore temporal trends by analyzing how revenue collection has changed over the months 2022.
monthly_revenue = df.groupby(df['CANCEL_DATE'].dt.to_period("M"))['TOTAL_COLLECTED'].sum()

# Plotting the temporal trends
plt.figure(figsize=(8, 4))
monthly_revenue.plot(marker='o', linestyle='-')
plt.title('Temporal Trends in Revenue Collection')
plt.xlabel('Month')
plt.ylabel('Total Collected Amount')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Revenue collection shows variations across months, with notable increases in February, April, and May. The municipality collected the highest revenue in April, contributing significantly to the overall trend.

```python
# 2. Are there particular months or quarters where the municipality tends to collect more revenue?

# Extracting month and quarter from the 'CANCEL_DATE' column
df['Month'] = df['CANCEL_DATE'].dt.month
df['Quarter'] = df['CANCEL_DATE'].dt.quarter

# Grouping by month and quarter, and summing the total collected amount for each
monthly_trends = df.groupby('Month')['TOTAL_COLLECTED'].sum()
quarterly_trends = df.groupby('Quarter')['TOTAL_COLLECTED'].sum()

# Plotting the temporal trends
plt.figure(figsize=(16, 6))

plt.subplot(1, 2, 1)
monthly_trends.plot(kind='bar', color='skyblue')
plt.title('Monthly Trends in Revenue Collection')
plt.xlabel('Month')
plt.ylabel('Total Collected Amount')

plt.subplot(1, 2, 2)
quarterly_trends.plot(kind='bar', color='lightcoral')
plt.title('Quarterly Trends in Revenue Collection')
plt.xlabel('Quarter')
plt.ylabel('Total Collected Amount')

plt.tight_layout()
plt.show()
```

Revenue collection tends to be higher in the second quarter, emphasizing the importance of analyzing quarterly trends in addition to monthly variations.

## Correlation Analysis:

```python
# Analyze correlations between different variables to understand relationships.
# Select relevant columns for correlation analysis
correlation_columns = ['TOTAL_COLLECTED', 'AMOUNT_C', 'RIGHT/LAW', 'INTERESTS_C', 'READJUSTMENT_C', 'DISCOUNT_C',
                        'RECEIVED', 'RETURNED', 'AMOUNT_D', 'RIGHT_D',
                        'INTEREST_D', 'READJUSTMENT_D', 'DISCOUNT_D', 'SUBTOTAL_D']

# Create a correlation matrix
correlation_matrix = df[correlation_columns].corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(8, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

Correlations between different financial variables indicate strong positive relationships. Notably, the total amount collected is highly correlated with various components, such as the amount charged, received, and specific financial elements like rights, interests, and discounts.

## Tax Concepts and Payments:

```python
# 1. Which tax concepts are more prevalent among the residents?

# Count the occurrences of each tax concept
tax_concept_counts = df['CONCEPT'].value_counts().head(5)

# Plot the distribution of tax concepts
plt.figure(figsize=(14, 6))
sns.barplot(x=tax_concept_counts.index, y=tax_concept_counts.values, palette='pastel')
plt.title('Distribution of Tax Concepts')
plt.xlabel('Tax Concept')
plt.ylabel('Number of Occurrences')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Municipal security, street sweeping, parks and gardens, garbage collection, and property tax are the most prevalent tax concepts among residents.

```python
# 2. What are the top concepts of payment and their corresponding collected amounts?

# Grouping by payment type and calculating total collected amount for each type
payment_type_amount = df.groupby('PAYMENT_TYPE')['TOTAL_COLLECTED'].sum().sort_values(ascending=False)

# Display the top concepts of payment and their corresponding collected amounts
print("Top Concepts of Payment and Corresponding Collected Amounts:")
print(payment_type_amount)
```

Transfers are the most common method of payment, contributing significantly to the highest total revenue. Cash transactions and VISA payments also contribute substantial amounts.

## Impact of Discounts:

```python
# How do offered discounts impact the total collected amount, and is there a noticeable correlation between discounts and timely payments?

# Scatter plot to visualize the relationship between discounts and total collected amount
plt.figure(figsize=(8, 4))
sns.scatterplot(x='DISCOUNT_D', y='TOTAL_COLLECTED', data=df)
plt.title('Impact of Discounts on Total Collected Amount')
plt.xlabel('Discount Amount')
plt.ylabel('Total Collected Amount')
plt.show()

# Correlation analysis
discount_total_corr = df['DISCOUNT_D'].corr(df['TOTAL_COLLECTED'])
print(f"Correlation between Discounts and Total Collected Amount: {discount_total_corr}")
```

The correlation between discounts and the total collected amount is positive but modest (0.137). While discounts have a measurable impact, their influence is not the sole driver of revenue collection.

## Interest Analysis:

```python
# Analyze the distribution of default interest and its contribution to total revenue
plt.figure(figsize=(8, 4))
sns.scatterplot(x='INTEREST_D', y='TOTAL_COLLECTED', data=df, color='red')
plt.title('Distribution of Default Interest and Its Contribution to Total Revenue')
plt.xlabel('Default Interest (Peruvian Sol)')
plt.ylabel('Total Revenue (Peruvian Sol)')
plt.show()
```

Default interest is concentrated in the lower range, with a long tail extending to higher values. Most values fall below 500 Peruvian Soles, suggesting that the majority of default cases are relatively minor.
The overall distribution of default interest is somewhat skewed to the right.
