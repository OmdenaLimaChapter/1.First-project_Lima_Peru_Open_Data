# Data Visualization

```python
%matplotlib inline
from scipy.stats import norm
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
```

```python
years = [2020, 2021, 2022]
total_2020 = debts_2020['TOTAL'].sum()
total_2021 = debts_2021['TOTAL'].sum()
total_2022 = debts_2022['TOTAL'].sum()

totals= [total_2020, total_2021, total_2022]
```

```python
# Create the figure and axis
plt.figure(figsize=(12, 6))
ax = plt.subplot(111)


# Create the bars for total cost
cost_bars = plt.bar(years, totals, color='red', label='Total Amount')

# Add value labels on top of the bars
for bar, value in zip(cost_bars, totals):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(value), ha='center', va='top', color='white')

plt.xlabel('Year')
plt.ylabel('Total Cost')
plt.title('Cost per Year')
plt.xticks(years)
plt.legend()
# plt.yticks(price_charged)
plt.figure(figsize=(15, 6))
plt.show()
```

```python
int_2020 = debts_2020['INTEREST'].sum()
int_2021 = debts_2021['INTEREST'].sum()
int_2022 = debts_2022['INTEREST'].sum()
int_sum = [int_2020, int_2021, int_2022]
```

```python
# Create the figure and axis
plt.figure(figsize=(12, 6))
ax = plt.subplot(111)


# Create the bars for total cost
cost_bars = plt.bar(years, totals, color='blue', label='Interest')

# Add value labels on top of the bars
for bar, value in zip(cost_bars, int_sum):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(value), ha='center', va='top', color='white')

plt.xlabel('Year')
plt.ylabel('Interest')
plt.title('Interest per Year')
plt.xticks(years)
plt.legend()
# plt.yticks(price_charged)
plt.figure(figsize=(15, 6))
plt.show()
```

```python
unpaid_2020 = debts_2020['UNPAID'].sum()
unpaid_2021 = debts_2021['UNPAID'].sum()
unpaid_2022 = debts_2022['UNPAID'].sum()
unpaid_sum = [unpaid_2020, unpaid_2021, unpaid_2022]
```

```python
# Create the figure and axis
plt.figure(figsize=(12, 6))
ax = plt.subplot(111)


# Create the bars for total cost
cost_bars = plt.bar(years, unpaid_sum, color='blue', label='Interest')

# Add value labels on top of the bars
for bar, value in zip(cost_bars, int_sum):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(value), ha='center', va='top', color='white')

plt.xlabel('Year')
plt.ylabel('Unpaid Amount')
plt.title('Unpaid Amount per Year')
plt.xticks(years)
plt.legend()
# plt.yticks(price_charged)
plt.figure(figsize=(15, 6))
plt.show()
```

```python
admin_2020 = debts_2020['ADMIN_EXPENSES'].sum()
admin_2021 = debts_2021['ADMIN_EXPENSES'].sum()
admin_2022 = debts_2022['ADMIN_EXPENSES'].sum()
admin_sum = [admin_2020, admin_2021, admin_2022]
admin_sum
```

```python
# Create the figure and axis
plt.figure(figsize=(12, 6))

# Use plot() to create the line graph, connecting points with lines
plt.plot(years, admin_sum, color='blue', label='Admin Expenses')

# Add value labels next to points for clarity
for year, value in zip(years, admin_sum):
    plt.annotate(str(value), xy=(year, value), ha='center', va='center')  # Use annotate for better placement

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Admin Expenses')
plt.title('Admin Expenses per Year (Line Graph)')
plt.xticks(years)
plt.legend()

# Adjust the figure size for better visual appeal
plt.figure(figsize=(10, 6))  # Changed for better balance
plt.show()
```

```python
readjustment_2020 = debts_2020['READJUSTMENT'].sum()
readjustment_2021 = debts_2021['READJUSTMENT'].sum()
readjustment_2022 = debts_2022['READJUSTMENT'].sum()
readjustment_sum = [readjustment_2020, readjustment_2021, readjustment_2022]
```

```python
import matplotlib.pyplot as plt

# Create the figure and axis
plt.figure(figsize=(12, 6))

# Use plot() to create the line graph, connecting points with lines
plt.plot(years, readjustment_sum, color='green', label='Readjustment')

# Add value labels next to points for clarity
for year, value in zip(years, readjustment_sum):
    plt.annotate(str(value), xy=(year, value), ha='center', va='center')  # Use annotate for better placement

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Readjustment')
plt.title('Readjustment per Year (Line Graph)')
plt.xticks(years)
plt.legend()

# Adjust the figure size for better visual appeal
plt.figure(figsize=(10, 6))  # Changed for better balance
plt.show()
```

```python
import matplotlib.pyplot as plt

# Combine the data for each line
data_to_plot = {
    "Readjustment": (years, readjustment_sum, "green"),
    "Unpaid Amount": (years, unpaid_sum, "blue"),
    "Interest": (years, totals, "purple"),  # Assuming "totals" holds interest data
    "Total Amount": (years, totals, "red"),
}

# Create the figure and axis
plt.figure(figsize=(12, 6))  # Adjust size as needed

# Plot each line with labels and customization
for label, (years, values, color) in data_to_plot.items():
    plt.plot(years, values, label=label, color=color)

# Add annotations for clarity (optional)
for year, value in zip(years, readjustment_sum):
    plt.annotate(str(value), xy=(year, value), ha='center', va='center')

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Amount')  # General label for combined values
plt.title('Multiple Financial Measures per Year (Line Graph)')
plt.xticks(years)
plt.legend()

plt.show()
```

```python
arbitrios_2020 = debts_2020.loc[debts_2020["TRIBUTE_DESC"] == "Arbitrios"].shape[0]
arbitrios_2021 = debts_2021.loc[debts_2021["TRIBUTE_DESC"] == "Arbitrios"].shape[0]
arbitrios_2022 = debts_2022.loc[debts_2022["TRIBUTE_DESC"] == "Arbitrios"].shape[0]
arbitrios_totals = [arbitrios_2020, arbitrios_2021, arbitrios_2022]
arbitrios_totals
```

```python
predial_2020 = debts_2020.loc[debts_2020["TRIBUTE_DESC"] == "Predial"].shape[0]
predial_2021 = debts_2021.loc[debts_2021["TRIBUTE_DESC"] == "Predial"].shape[0]
predial_2022 = debts_2022.loc[debts_2022["TRIBUTE_DESC"] == "Predial"].shape[0]
predial_totals = [predial_2020, predial_2021, predial_2022]
predial_totals
```

```python
multatribu_2020 = debts_2020.loc[
    (debts_2020["TRIBUTE_DESC"] == "Multa Tributaria") |
    (debts_2020["TRIBUTE_DESC"] == "Multas Tributarias")
].shape[0]
multatribu_2021 = debts_2021.loc[
    (debts_2021["TRIBUTE_DESC"] == "Multa Tributaria") |
    (debts_2021["TRIBUTE_DESC"] == "Multas Tributarias")
].shape[0]
multatribu_2022 = debts_2022.loc[
    (debts_2022["TRIBUTE_DESC"] == "Multa Tributaria") |
    (debts_2022["TRIBUTE_DESC"] == "Multas Tributarias")
].shape[0]
multatribu_totals = [multatribu_2020, multatribu_2021, multatribu_2022]
multatribu_totals
```

```python
multa_2020 = debts_2020.loc[debts_2020["TRIBUTE_DESC"] == "Multa Administrativa"].shape[0]
multa_2021 = debts_2021.loc[debts_2021["TRIBUTE_DESC"] == "Multa Admininstrativa"].shape[0]
multa_2022 = debts_2022.loc[debts_2022["TRIBUTE_DESC"] == "Multa Admininstrativa"].shape[0]
multa_totals = [multa_2020, multa_2021, multa_2022]
multa_totals
```

```python
bar_width = 0.2

# Create x-axis values for the bars
x = range(len(years))

# Create the bar chart
plt.bar(x, multa_totals, width=bar_width, label='Multas Administrativas', color='red', align='center')
plt.bar([i + bar_width for i in x], multatribu_totals, width=bar_width, label='Multas Tributarias', color='blue', align='center')
plt.bar([i + 2 * bar_width for i in x], predial_totals, width=bar_width, label='Predial', color='green', align='center')
plt.bar([i + 3 * bar_width for i in x], arbitrios_totals, width=bar_width, label='Arbitrios', color='purple', align='center')

# Set the x-axis labels and the legend
plt.xticks([i + 1.5 * bar_width for i in x], years)
plt.xlabel('Year')
plt.ylabel('Cases')
plt.title('Number of Cases Based on Tax Description')
plt.legend()

# Show the plot
plt.show()
```

```python

plt.figure(figsize=(12, 6))
ax = plt.subplot(111)

bar_admin = plt.bar(years, admin_sum, color='red', label='Admin Expenses')
bar_readjustment = plt.bar(years, readjustment_sum, color='blue', label='Readjustment', bottom=admin_sum)
bar_unpaid = plt.bar(years, unpaid_sum, color='green', label='Unpaid', bottom=[sum(x) for x in zip(admin_sum, readjustment_sum)])
bar_interest = plt.bar(years, int_sum, color='orange', label='Interest', bottom=[sum(x) for x in zip(admin_sum, readjustment_sum, unpaid_sum)])

# Add labels for all values, adjusting for stacked bars
for bar, value, bottom in zip(bar_admin + bar_readjustment + bar_unpaid + bar_interest, admin_sum + readjustment_sum + unpaid_sum + int_sum, [0] * len(years) + admin_sum + [sum(x) for x in zip(admin_sum, readjustment_sum)] + [sum(x) for x in zip(admin_sum, readjustment_sum, unpaid_sum)]):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + bottom, str(value), ha='center', va='bottom', color='black')  # Use 'bottom' for better placement

plt.xlabel('Year')
plt.ylabel('Total Cost')
plt.title('Total Expenses per Year')
plt.xticks(years)
plt.legend()

plt.show()
```
