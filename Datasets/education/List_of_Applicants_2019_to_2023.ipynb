```python
List_of_Applicants_all=pd.read_csv('/content/drive/MyDrive/Omdena/Education/LISTA_DE_POSTULANTES_1.csv',sep=';', encoding=('ISO-8859-1'))
List_of_Applicants=List_of_Applicants_all[List_of_Applicants_all['DEPARTAMENTO']=='LIMA']
print(List_of_Applicants.shape)
List_of_Applicants.head()
```

# DATA VISUALISATION

DEMOGRPHIC ANALYSIS

```python
# Calculate summary statistics for age
age_summary = List_of_Applicants['EDAD'].describe()

# Create a histogram for age distribution
plt.figure(figsize=(10, 6))
plt.hist(List_of_Applicants['EDAD'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Applicants')
plt.xlabel('Age')
plt.ylabel('Number of Applicants')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Display summary statistics
print("Age Summary Statistics:\n", age_summary)
```

'''The majority of applicants fall within the age range of 17 to 20, with an average age of approximately 19 years. The dataset includes a range of ages, but the majority of applicants are relatively young.'''

Educational Background

```python
school_counts = List_of_Applicants['ESCUELA'].value_counts()
plt.figure(figsize=(12, 6))
school_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of Applicants by School Type')
plt.xlabel('School Type')
plt.ylabel('Number of Applicants')
plt.xticks(rotation=45, ha='right')
plt.show()
```

The distribution of applicants by the type of school they attended indicates a diverse set of interests. Engineering programs, especially Civil Engineering, are popular choices, and there is also interest in fields related to the environment (Ecotourism), food (Food Engineering and Tropical Agronomic Engineering), economics, and accounting.

```python
educational_level_counts = List_of_Applicants['MODALIDAD'].value_counts()
plt.figure(figsize=(6, 6))
educational_level_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightgreen', 'lightskyblue'])
plt.title('Distribution of Applicants by Educational Level')
plt.ylabel('')
plt.show()
```

The majority of applicants are applying through the regular admission process (ORDINARIO). However, there are also applicants utilizing other admission pathways, including those designed for high school students in their final year, top-performing students, scholarship programs, and individuals with disabilities.

```python
# Distribution of Applicants with Disabilities
disability_counts = List_of_Applicants['TIENE_DISCAPACIDAD'].value_counts()

plt.figure(figsize=(6,6))
disability_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'])
plt.title('Distribution of Applicants with Disabilities')
plt.ylabel('')
plt.show()
```

The dataset suggests that disabilities are relatively uncommon among the applicants, with only one applicant reporting a disability. It's important to note that the dataset's size may impact the interpretation of these numbers, and the conclusion is based on the available data.

```python
# Number of Applicants Each Year
yearly_counts = List_of_Applicants['ANNIO_POSTULACION'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.plot(yearly_counts.index, yearly_counts.values, marker='o', linestyle='-', color='skyblue')
plt.title('Number of Applicants Each Year')
plt.xlabel('Year of Admission')
plt.ylabel('Number of Applicants')
plt.grid(True)
plt.show()
```

There is a positive trend in the number of applicants over the years, with a significant increase in 2023.

```python
language_proficiency_columns = ['IDIOMA', 'LEE', 'ESCRIBE', 'HABLA']

plt.figure(figsize=(12, 8))

for column in language_proficiency_columns:
    proficiency_counts = List_of_Applicants[column].value_counts()
    plt.subplot(2, 2, language_proficiency_columns.index(column) + 1)
    proficiency_counts.plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Number of Applicants')

plt.tight_layout()
plt.show()
```

Language of Applicants (IDIOMA):
The majority of applicants (59 individuals) indicate proficiency in Spanish ('Español').
There are 6 applicants who indicate proficiency in Quechua, and 1 applicant who indicates proficiency in Yine.

Reading Proficiency (LEE):
The vast majority of applicants (65 individuals) report that they can read ('SI').
Only 1 applicant reports that they cannot read ('NO').

Writing Proficiency (ESCRIBE):
The majority of applicants (54 individuals) report that they can write ('SI').
There are 12 applicants who report that they cannot write ('NO').

Speaking Proficiency (HABLA):
The majority of applicants (53 individuals) report that they can speak ('SI').
There are 13 applicants who report that they cannot speak ('NO')

```python
# Filter applicants who speak Quechua or Yine
quechua_yine_applicants = List_of_Applicants[(List_of_Applicants['IDIOMA'] == 'Quechua') | (List_of_Applicants['IDIOMA'] == 'Yine')]

# Display information about these applicants
# print("Applicants who speak Quechua or Yine:")
# print(quechua_yine_applicants)
```

```python
# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,6))

# Age Distribution
sns.histplot(data=quechua_yine_applicants, x='EDAD', bins=15, kde=True, color='skyblue', ax=axes[0])
axes[0].set_title('Age Distribution of Quechua and Yine Speaking Applicants')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Number of Applicants')

# Gender Distribution
gender_counts = quechua_yine_applicants['SEXO'].value_counts()
gender_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightblue'], ax=axes[1])
axes[1].set_title('Gender Distribution of Quechua and Yine Speaking Applicants')
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()
```

The applicants who speak Quechua or Yine exhibit a diverse age distribution, with a concentration in the late teens and early twenties.

The gender distribution is slightly skewed towards males, with a majority of 57.1%.

These insights provide a glimpse into the demographic characteristics of applicants who speak Quechua or Yine.

```python

# Explore geographical distribution (e.g., UBIGEO, DEPARTAMENTO, PROVINCIA, DISTRITO)
geo_distribution = quechua_yine_applicants[['UBIGEO', 'DEPARTAMENTO', 'PROVINCIA', 'DISTRITO']]
print("\nGeographical Distribution:")
print(geo_distribution)

# Explore educational background (e.g., ESCUELA, MODALIDAD)
edu_background = quechua_yine_applicants[['ESCUELA', 'MODALIDAD']]
# print("\nEducational Background:")
# print(edu_background)
```

```python
# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 12))

# Geographical Distribution
geo_distribution = quechua_yine_applicants[['DEPARTAMENTO', 'PROVINCIA', 'DISTRITO']]
sns.countplot(x='DEPARTAMENTO', data=geo_distribution, ax=axes[0, 0])
axes[0, 0].set_title('Geographical Distribution by Department')
axes[0, 0].set_ylabel('Number of Applicants')
axes[0, 0].set_xlabel('Department')

sns.countplot(x='PROVINCIA', data=geo_distribution, ax=axes[0, 1])
axes[0, 1].set_title('Geographical Distribution by Province')
axes[0, 1].set_ylabel('Number of Applicants')
axes[0, 1].set_xlabel('Province')

sns.countplot(x='DISTRITO', data=geo_distribution, ax=axes[1, 0])
axes[1, 0].set_title('Geographical Distribution by District')
axes[1, 0].set_ylabel('Number of Applicants')
axes[1, 0].set_xlabel('District')
axes[1, 0].set_xticklabels(axes[1, 0].get_xticklabels(), rotation=90, ha='right')

# Educational Background
edu_background = quechua_yine_applicants[['ESCUELA', 'MODALIDAD']]
sns.countplot(x='ESCUELA', data=edu_background, ax=axes[1, 1])
axes[1, 1].set_title('Educational Background by School Type')
axes[1, 1].set_ylabel('Number of Applicants')
axes[1, 1].set_xlabel('School Type')
```

The geographical distribution of the provided applicants who speak Quechua or Yine is concentrated in the department, province, and districts within Lima.
