```python
def plot_top_bottom_courses(data, year):
    # Filter data for the specific year
    year_data = data[data['ANNIO'] == year]

    # Top 5 courses
    top_courses = year_data.groupby('CURSO1')['CURSO1'].count().sort_values(ascending=False).head(5).index

    # Bottom 5 courses
    bottom_courses = year_data.groupby('CURSO1')['CURSO1'].count().sort_values().head(5).index

    # Combine data for top and bottom courses
    combined_courses = pd.concat([year_data[year_data['CURSO1'].isin(top_courses)],
                                 year_data[year_data['CURSO1'].isin(bottom_courses)]])

    # Create subplot
    plt.figure(figsize=(6,4))

    # Plot Top 5
    plt.subplot(1, 2, 1)
    sns.countplot(data=combined_courses[combined_courses['CURSO1'].isin(top_courses)],
                  x='CURSO1', order=top_courses, palette='viridis')
    plt.title(f'Top 5 Courses in {year}')
    plt.xticks(rotation=45, ha='right')

    # Plot Bottom 5
    plt.subplot(1, 2, 2)
    sns.countplot(data=combined_courses[combined_courses['CURSO1'].isin(bottom_courses)],
                  x='CURSO1', order=bottom_courses, palette='magma')
    plt.title(f'Bottom 5 Courses in {year}')
    plt.xticks(rotation=45, ha='right')

    plt.tight_layout()
    plt.show()

# Assuming the dataset contains years from 2019 to 2022
for year in range(2019, 2023):
    plot_top_bottom_courses(filtered_lima_list, year)
```

'''Top 5 Courses Trends:
Courses such as MATEMÁTICA I, TÉCNICAS, and ETNOLOGÍA consistently appear among the top 5 courses across multiple years.
Stability in the popularity of certain courses suggests a continued interest or requirement in these subjects.

Bottom 5 Courses Trends:
The courses in the bottom 5 exhibit fluctuations across the years, indicating varying levels of enrollment or interest.
Some courses may experience lower enrollment consistently, while others may have experienced fluctuations in popularity.

Yearly Variations:
Observing the courses that enter or exit the top and bottom 5 lists each year provides insights into changing trends in student preferences or academic requirements.
