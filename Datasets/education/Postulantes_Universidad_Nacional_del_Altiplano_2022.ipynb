#### Convert date columns for YYYY/MM/DD format

**Columns:**
 - `FECHA_CORTE`
 - `F_NACIMIENTO`

```python
lima_df.head()
```

```python
def convert_date(serie):
    dates = []
    for value in serie.reset_index(drop=True):
        date_string = str(int(value))
        year, month, day = date_string[:4],date_string[4:6], date_string[6:]
        year, month, day = int(year), int(month), int(day)
        date_value = date(year, month, day)
        dates.append(date_value)

    return dates
```

```python
lima_df.FECHA_CORTE = convert_date(lima_df.FECHA_CORTE)
lima_df.F_NACIMIENTO = convert_date(lima_df.F_NACIMIENTO)
```
