### Preprocessing

#### Missing values

```python
lima_df.isnull().sum()
```

```python
average_age = lima_df.EDAD.mean()
print(f'Average age: {average_age}')
```

```python
age_values = lima_df.EDAD.copy()
lima_df.EDAD = age_values.fillna(value=average_age)
```

```python
birthday_mising_value = lima_df[lima_df.EDAD == np.ceil(average_age)].FECHA_NACIMIENTO.mode().values[0]
print(f'Birthday values for the missing value: {birthday_mising_value}')
```

```python
birthday_values = lima_df.FECHA_NACIMIENTO.copy()
lima_df.FECHA_NACIMIENTO = birthday_values.fillna(value=birthday_mising_value)
```

```python
lima_df.isnull().sum()
```

#### Convert date columns for YYYY/MM/DD format

**Columns:**
 - `FECHA_CORTE`
 - `FECHA_NACIMIENTO`
 - `FECHA_INSCRIP`

```python
lima_df.head()
```

```python
def convert_date(serie):
    dates = []
    for value in serie.reset_index(drop=True):
        date_string = str(int(value))
        year, month, day = date_string[:4],date_string[4:6], date_string[6:]
        year, month, day = int(year), int(month), int(day)
        date_value = date(year, month, day)
        dates.append(date_value)

    return dates
```

```python
lima_df.FECHA_CORTE = convert_date(lima_df.FECHA_CORTE)
lima_df.FECHA_NACIMIENTO = convert_date(lima_df.FECHA_NACIMIENTO)
lima_df.FECHA_INSCRIP = convert_date(lima_df.FECHA_INSCRIP)
```
