# New Oncology Patients – [National Institute of Neoplastic Diseases - INEN]

## Dataset Overview:

### This dataset comprises daily records of new oncology patients, where each record corresponds to an individual. 

### Each entry is characterized by key details, including the primary department of care, date of affiliation, gender, age, and the geographical locations of both the place of birth and residence, detailed up to the district level. 

### The dataset comprises 4,658 entries and includes 9 features capturing information about different aspects of patients profiles.

### Access the Dataset: https://www.datosabiertos.gob.pe/dataset/pacientes-oncol%C3%B3gicos-nuevos-%E2%80%93-instituto-nacional-de-enfermedades-neopl%C3%A1sicas-inen-0

```python
# Creating a histogram using Seaborn to visualize the distribution of ages

# Setting up the figure size for the plot
plt.figure(figsize=(10, 5))

# Creating a histogram
sns.histplot(data=df, x='EDAD', bins=100, kde=True)

# Adding title and labels to the plot
plt.title('Distribución de edad')
plt.xlabel('Edad')
plt.ylabel('Número de Pacientes')

# Displaying the plot
plt.show()
```

### In summary, our analysis of the oncology dataset uncovers notable trends in age distribution. The histogram peaks around mid-sixties, suggesting a concentration of cases in that age group. Additionally, a distinct rise around age 40 and a subsequent decline after age 80 are observed. Average patient counts reveal increasing trends with age, providing valuable insights for tailoring targeted interventions in oncology.

```python
# Setting up the figure size for the pie chart
plt.figure(figsize=(6, 6))

# Creating a pie chart to visualize the distribution of 'SEXO' (gender)
df['SEXO'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=0, wedgeprops=dict(width=0.7))

# Adding title and removing y-axis label
plt.title('Distribución de género')
plt.ylabel('')

# Displaying the pie chart
plt.show()
```

### In the examination of gender distribution within the oncology dataset, the pie chart reveals a significant majority of female patients, comprising 64.7% of the total, while the remaining percentage represents male patients. This gender disproportion suggests a potential gender-related aspect in the prevalence of oncological conditions. Understanding such disparities is crucial for tailoring healthcare strategies to address the specific needs of different genders within the context of oncology.

```python
# Setting up the figure size for the histogram with gender classification
plt.figure(figsize=(10, 5))

# Defining colors for the gender categories
colors = {'MASCULINO':'orange', 'FEMENINO':'yellow'}

# Creating a stacked histogram with Seaborn to visualize age distribution by gender
sns.histplot(data=df, x='EDAD', hue='SEXO', multiple='stack', bins=100, kde=True, palette=colors)

# Adding title and labels to the plot
plt.title('Distribución de edad por género')
plt.xlabel('Edad')
plt.ylabel('Número de Pacientes')

# Displaying the plot
plt.show() 
```

### In our exploration of age and gender distribution in the oncology dataset, the histogram offers technical insights. The plot reveals a peak of 100 female patients at age 65, forming a distinctive bell curve. Simultaneously, for males, there's a peak of 40 patients centered around age 70, exhibiting a similar bell-shaped distribution. This graphical representation underscores the nuanced interplay between age and gender, providing a valuable technical perspective on the specific age ranges where cancer cases concentrate for both genders.

```python
# Creating a count plot to visualize the value counts of 'DEPARTAMENTO_PRINCIPAL'
sns.countplot(data=df, x='DEPARTAMENTO_PRINCIPAL', palette='turbo')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45,ha='right')

# Adding title and labels to the plot
plt.title('Conteo de Pacientes por Departamento de Oncología')
plt.xlabel('Departamento')
plt.ylabel('Número de Pacientes')

# Displaying the plot
plt.show()
```

### In summary, our analysis of oncology department data reveals varied patient counts across different areas. ABDOMEN, GINECOLOGIA, and MAMAS Y TEJIDOS BLANDOS lead in cases, suggesting diverse cancer prevalence. ONCOLOGIA PEDIATRICA and MEDICINA NUCLEAR, with fewer cases, underscore specialized challenges. This insight aids in tailoring healthcare strategies for distinct oncological needs.

```python
# Creating subplots with a specified figure size
fig, ax = plt.subplots(figsize=(16, 8))

# Defining colors for different oncology departments
colors = {'ABDOMEN': 'Red', 'GINECOLOGIA': 'Green', 'MAMAS Y TEJIDOS BLANDOS': 'Blue',
          'CABEZA Y CUELLO': 'Turquoise', 'UROLOGIA': 'Purple', 'MEDICINA': 'Orange',
          'RADIOTERAPIA': 'Teal', 'TORAX': 'Pink', 'NEUROCIRUGIA': 'Brown',
          'ONCOLOGIA PEDIATRICA': 'Gray', 'MEDICINA NUCLEAR': 'Cyan', 'ORTOPEDIA ONCOLOGICA': 'Magenta'}

# Creating a kernel density plot to visualize age distribution by different oncology departments
sns.kdeplot(x='EDAD', hue='DEPARTAMENTO_PRINCIPAL', data=df, palette=colors)

# Setting title for the plot
ax.set(title='Distribución de Edades en Diferentes Departamentos de Oncología')
ax.set_xlabel('Edad')

# Displaying the plot
plt.show()
```

### In the realm of oncology departments, ABDOMEN emerges as a focal point, harboring a high density of patients primarily between the ages of 65-75. MAMAS Y TEJIDOS BLANDOS follows closely, displaying a slightly lower density within the age bracket of 40-50. GINECOLOGIA, with patients ranging from 40-60, also contributes to this diverse landscape. Notably, ONCOLOGIA PEDIATRICA, dealing with younger patients aged 5-20, exhibits a lower density. This nuanced age-specific distribution underscores the intricate relationship between cancer types and patient demographics within each department.
