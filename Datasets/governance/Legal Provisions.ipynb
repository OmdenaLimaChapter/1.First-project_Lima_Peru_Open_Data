# Legal Devices

Legal Devices published in the Official Gazette El Peruano, which are issued by State Entities. These can be laws, resolutions, decrees, agreements, etc.  

This DataSet details:  

Publication Date, Publication Order, Entity, Device Type, Number, Summary, Link, Cut-Off Date.  

Legal devices are shown from 2013 to March 2023

### Data Source

https://www.datosabiertos.gob.pe/dataset/dispositivos-legales

### Combine dataframes into one

```python
# List of dataframes to concatenate
dataframes = [df_2013_2022, df_2022, df_2023_01, df_2023_02, df_2023_03]

# Concatenate dataframes vertically
df_combined = pd.concat(dataframes, ignore_index=True)
```

Check number of rows.

```python
row_count = 0
for df in dataframes:
    print(df.shape[0])
    row_count += df.shape[0]
    
row_count
```

### Remove duplicates

Remove duplicates and check number of rows

```python
df_no_duplicates = df_combined.drop_duplicates(subset='OP', keep='first')
df_no_duplicates.shape # Expected number of rows is 209892
```

### Exploratory Data Analysis

```python
df_no_duplicates.columns
```

'PUBLICATION DATE', 'OP', 'ENTITY', 'DEVICE', 'NUMBER', 'TIP', 'LINK', 'CUT-OFF DATE'

- PUBLICATION DATE
Publication date of the legal device - Numeric 8 Format: yyyymmdd

- OP
Publication Order Alphanumeric 20

- ENTITY
Name of the entity generating the legal device Alphanumeric 255

- DEVICE
Device typing Alphanumeric 255

- NUMBER
Device identification Alphanumeric 1024 

- SUMILLA 
Text describing the contents of the device Alphanumeric 2048

- LINK
Link to download the device in pdf format published in elperuano.pe Alphanumeric 250

- CUT_DATE
Date on which the dataset was generated Numeric 8 Format: yyyymmdd

```python
df_no_duplicates.head(30)
```

```python
# Assuming df_no_duplicates is your DataFrame
df_no_duplicates = df_no_duplicates.copy()

# Convert 'FECHA_PUBLICACION' to string
df_no_duplicates['FECHA_PUBLICACION'] = df_no_duplicates['FECHA_PUBLICACION'].astype(str)

# Create new columns 'pub_year' and 'pub_month'
df_no_duplicates['pub_year'] = df_no_duplicates['FECHA_PUBLICACION'].str[:4].astype(int)
df_no_duplicates['pub_month'] = df_no_duplicates['FECHA_PUBLICACION'].str[4:6].astype(int)
```

```python
df_no_duplicates.head()
```

```python
df_no_duplicates['pub_year'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Publications by Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
df_no_duplicates['DISPOSITIVO'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Publications by Entity')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
